```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
d2w10_train = np.load('Data/2A/D2W10/train_losses.npy')
d2w10_test = np.load('Data/2A/D2W10/test_losses.npy')

d3w160_train = np.load('Data/2A/D3W160/train_losses.npy')
d3w160_test = np.load('Data/2A/D3W160/test_losses.npy')

d4w2560_train = np.load('Data/2A/D4W2560/train_losses.npy')
d4w2560_test = np.load('Data/2A/D4W2560/test_losses.npy')
```

```python
iters = np.arange(len(d2w10_test))
skip = 400
tick_skip = 1000

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(iters[::skip], d2w10_train[::skip], 'r', label = 'D2W10:Train')        
ax.plot(iters[::skip], d2w10_test[::skip], 'r--', label = 'D2W10:Test')

ax.plot(iters[::skip], d3w160_train[::skip], 'b', label = 'D3W160:Train')        
ax.plot(iters[::skip], d3w160_test[::skip], 'b--', label = 'D3W160:Test')


#ax.plot(iters[::skip], d4w2560_train[::skip], 'g', label = 'D4W2560:Train')        
#ax.plot(iters[::skip], d4w2560_test[::skip], 'g--', label = 'D4W2560:Test')

plt.xticks(iters[::tick_skip])
plt.xlabel('# Iterations')
plt.ylabel('MSE')
plt.legend()

plt.ylim([ -10**(-2), 10**(-1)])
plt.savefig('MSE-vs-Iterations.jpg', dpi = 500)



plt.show()
```
